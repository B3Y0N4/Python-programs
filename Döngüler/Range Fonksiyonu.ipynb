# Range Fonksiyonu Kullanımı

```python
range(0,20)
```

```python
print(range(0,20))
```

```python
print(*range(0,20))
```

```python
print(*range(20))
```

```python
print(*range(5,10,2))
```

```python
print(*range(5,100,5))
```

```python
print(*range(100,0))
```

```python
print(*range(100,0,-1))
```

# Range Fonksiyonunun For Döngüsüyle Kullanımı

```python
for i in range(0,99):
    print(i)
```

```python
liste = []
```

```python
for i in range(101):
    liste.append(i)
print(liste)
```

```python
for i in range(0,11):
    print(i*"*")
```

```python
for i in range(0,100):
    print(i*"*")
```
